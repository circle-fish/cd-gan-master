```python
import pandas as pd
import os
import os.path as path
from clickhouse_driver import Client
```

```python
client = Client('localhost')
```

```python
client.execute('SELECT COUNT(*) FROM ship.type')
```

```python
result = client.execute('''
        SELECT mmsi_hash, ship_type_code
        from ship.tank_sp_f group by mmsi_hash,ship_type_code
        order by ship_type_code, mmsi_hash
''')
```

```python
pd.DataFrame(result, columns=['mmsi_hash', 'ship_type_code'])
```

```python
result = client.execute('''
        SELECT flag_hash,  count( distinct primary_vessel_type)
        from ship.tank_sp_f group by flag_hash
        order by flag_hash
''')
```

```python
pd.DataFrame(result, columns=['flag_hash','primary_vessel_type'])
```

```python
result = client.execute("""
                        SELECT 
                        distinct primary_vessel_type, ui_vessel_type
                        FROM ship.type 
                        where  primary_vessel_type in ('BULK CARRIER', 'PASSENGER','TANKER', 'UTILITY', 'SPECIALIST')                         
                        order by primary_vessel_type, ui_vessel_type
                        """)
result = pd.DataFrame(result)

"""

 columns=['data_source_reference',
                               'primary_vessel_type',
                               'specific_vessel_type',
                               'ui_vessel_type',
                               'data_source_raw_vessel_type']
                               """
```

```python
result = client.execute("""
                        SELECT 
                        distinct data_source_reference
                        FROM ship.type 
                        where  type.primary_vessel_type in ('TANKER', 'SPECIALIST') 
                        and type.ui_vessel_type in ('BARGE', 'SPECIALIST VESSEL', 'TANKER')  
                        """)

```

```python
st = "where ship_type_code in ("
for r in result:
    st = st + "'" + r[0] + "',"
print(st)
```

```python
result = client.execute("""
                        SELECT 
                        tank_sp.*, type.primary_vessel_type, type.ui_vessel_type
                        FROM ship.tank_sp join ship.type on type.data_source_reference = tank_sp.ship_type_code                        
                        """)
```

```python
result
```

```python
abc = pd.DataFrame(result, columns=['time_utc',
    'longitude',
    'latitude',
    'mmsi_hash',
    'speed',
    'heading',    
    'flag_hash',
    'breadth_extream',
    'draught_large',
    'grt',
    'dwt',    
    'ship_type_code',
    'filename',
    'primary_vessel_type',
    'ui_vessel_type'])
```
