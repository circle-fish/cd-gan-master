```python
data_dir = path.join(base_dir, main_data_dir)
```

```python
csv_files = [x for x in os.listdir(data_dir) if path.isfile(path.join(data_dir, x))]
```

```python
csv_files
```

```python
data_df = pd.read_csv(path.join(data_dir, csv_files[0]), sep=',')
```

```python
wanted_columns = ['time_utc',
                  'longitude',
                  'latitude',
                  'mmsi_hash',
                  'speed',
                  'heading',
                  'flag_hash',
                  'breadth_extream',
                  'draught_large',
                  'grt',
                  'dwt',
                  'ship_type_code',
                  ]
```

```python
for f in csv_files:
    data_df = pd.read_csv(path.join(data_dir, f), sep=',')    
    selected_df = data_df[wanted_columns]
    selected_df['filename'] = f
    selected_df.to_csv(path.join(base_dir, 'selected-data',f), header=False, index=False)
```

```python
for f in csv_files:
    data_df = pd.read_csv(path.join(data_dir, f), sep=',')    
    selected_df = data_df[wanted_columns]
    selected_df['filename'] = f
    selected_df.to_csv(path.join(base_dir, 'selected-data',f))
```
