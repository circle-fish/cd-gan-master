## Preprocessing

```python
prim_vessel_cat = pd.api.types.CategoricalDtype(categories=['TANKER','SPECIALIST'], ordered=False)
ui_vessel_cat = pd.api.types.CategoricalDtype(categories=['BARGE', 'SPECIALIST VESSEL', 'TANKER'], ordered=False)
                                                            
```

```python
min_max_cols = ['breadth_extream','draught_large','grt','dwt','speed']
type_list = min_max_cols + ['longitude','latitude','heading']
d_16_map = {x:np.float32 for x in type_list}
df = df.astype(d_16_map, copy=False)
prepro = MinMaxScaler()
df.loc[:,min_max_cols] = prepro.fit_transform(df[min_max_cols])
df['prim']=df['primary_vessel_type'].astype(prim_vessel_cat)
df['ui'] = df['ui_vessel_type'].astype(ui_vessel_cat)
prim = pd.get_dummies(df['prim'], prefix='prim')        
ui = pd.get_dummies(df['ui'], prefix='ui')  
df = df.astype(d_16_map, copy=False)
```

```python
df.dtypes
```

```python
result = pd.concat([df,prim,ui], axis=1)
```

```python
result.drop(['primary_vessel_type','ui_vessel_type','prim','ui'],axis=1,inplace=True)
```

```python
result.drop(['ship_type_code'],axis=1,inplace=True)
```

```python
result.to_csv(path.join(base_dir,'tk_sp_trainable.csv'), index=False)
```
